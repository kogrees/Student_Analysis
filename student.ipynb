```python
import numpy as np
import pandas as pd
```

```python
url = "C:/Users/kogre/Downloads/project/student/StudentPerformanceFactors.csv"
```

```python
df = pd.read_csv(url, header = None)
```

```python
df.columns = df.iloc[0]
```

```python
df.info()
```

```python
df.head()
```

```python
df.drop([0], inplace = True)
```

```python
df.head()
```

```python
df[['Hours_Studied','Attendance','Sleep_Hours','Previous_Scores','Tutoring_Sessions','Physical_Activity','Exam_Score']] = df[['Hours_Studied','Attendance','Sleep_Hours','Previous_Scores','Tutoring_Sessions','Physical_Activity','Exam_Score']].astype(int)
```

```python
df.info()
```

```python
df.isnull().any()
```

```python
mode_TQ = df["Teacher_Quality"].mode()[0]
df['Teacher_Quality'] = df['Teacher_Quality'].replace(np.NaN, mode_TQ)
```

```python
print(len(df))
```

```python
df.dropna(subset = "Parental_Education_Level", inplace = True)
df.dropna(subset = "Distance_from_Home", inplace = True)
```

```python
df.isnull().any()
```

```python
print(len(df))
```

```python
df.head(10)
```

```python
capitilazion_df = ['Parental_Involvement','Access_to_Resources','Extracurricular_Activities','Motivation_Level','Internet_Access','Family_Income','Teacher_Quality','School_Type','Peer_Influence','Learning_Disabilities','Parental_Education_Level','Distance_from_Home','Gender']
for column in capitilazion_df:
    df[column] = df[column].str.capitalize().str.strip()

df
```

```python
df.to_csv('student_analysis.csv')
```
